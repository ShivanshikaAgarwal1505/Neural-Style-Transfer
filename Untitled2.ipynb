```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
# Image size (keep small for speed)
IMAGE_SIZE = 256

loader = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor()
])

def load_image(image_path):
    image = Image.open(image_path).convert("RGB")
    image = loader(image).unsqueeze(0)  # (1, 3, H, W)
    return image.to(device)
```

```python
# load image

content_img = load_image("content.jpg")
style_img = load_image("style.jpg")

print(content_img.shape, style_img.shape)
```

```python
def normalize(img):
    mean = img.mean(dim=[2,3], keepdim=True)
    std = img.std(dim=[2,3], keepdim=True) + 1e-5
    return (img - mean) / std

content_img = normalize(content_img)
style_img = normalize(style_img)
```

```python
# visualize image

def imshow(tensor, title=None):
    image = tensor.clone().detach().cpu().squeeze(0)
    image = image.permute(1, 2, 0)
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()

imshow(content_img, "Content Image")
imshow(style_img, "Style Image")
```

```python
# CNN

class BetterCNN(nn.Module):
    def __init__(self):
        super(BetterCNN, self).__init__()

        self.layers = nn.ModuleList([
            nn.Conv2d(3, 16, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.AvgPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
        ])
    def forward(self, x):
      features = []
      for layer in self.layers:
          x = layer(x)
          if isinstance(layer, nn.Conv2d):
              features.append(x)
      return features
```

```python
# model initialze

model = BetterCNN().to(device)
```

```python
# extract feature maps

content_features = model(content_img)
style_features = model(style_img)
```

```python
# inspect feature shapes

for i, feat in enumerate(content_features):
    print(f"Layer {i} feature shape: {feat.shape}")
```

```python
# visualize feature maps

def show_feature_maps(feature_tensor, num_maps=6):
    feature_tensor = feature_tensor[0]  # remove batch
    plt.figure(figsize=(15, 5))

    for i in range(num_maps):
        plt.subplot(1, num_maps, i + 1)
        plt.imshow(feature_tensor[i].detach().cpu(), cmap="gray")
        plt.axis("off")

    plt.show()
```

```python
show_feature_maps(content_features[0])   # shallow
show_feature_maps(content_features[-1])  # deep
```

```python
# freeze CNN

for param in model.parameters():
    param.requires_grad = False
```

```python
# Initialize the Generated Image

generated_img = (
    0.7 * content_img +
    0.3 * torch.randn_like(content_img)
).clamp(0, 1).requires_grad_(True)
```

```python
# Choose Content & Style Layers

CONTENT_LAYER = 2
STYLE_LAYERS = [0, 1, 2]
```

```python
# Content Loss

def content_loss(gen_feat, content_feat):
    return torch.mean((gen_feat - content_feat) ** 2)
```

```python
# Gram Matrix

def gram_matrix(feature):
    b, c, h, w = feature.shape
    feature = feature.view(c, h * w)
    gram = torch.mm(feature, feature.t())
    return gram / (c * h * w)
```

```python
# Style Loss

def style_loss(gen_features, style_features):
    loss = 0
    for i in STYLE_LAYERS:
        G = gram_matrix(gen_features[i])
        A = gram_matrix(style_features[i])
        loss += torch.mean((G - A) ** 2)
    return loss
```

```python
# Total Loss

alpha = 10       # content weight
beta = 200     # style weight (CPU-friendly)

tv_weight = 5e-4

def total_loss(gen_features, content_features, style_features, img):
    c_loss = content_loss(
        gen_features[CONTENT_LAYER],
        content_features[CONTENT_LAYER]
    )
    s_loss = style_loss(gen_features, style_features)
    tv_loss = total_variation_loss(img)

    return alpha * c_loss + beta * s_loss + tv_weight * tv_loss
```

```python
def total_variation_loss(img):
    loss = torch.mean(torch.abs(img[:, :, :, :-1] - img[:, :, :, 1:])) + \
           torch.mean(torch.abs(img[:, :, :-1, :] - img[:, :, 1:, :]))
    return loss
```

```python
# Optimizer

optimizer = optim.Adam([generated_img], lr=0.01)
```

```python
# Extract content & style features ONCE (no gradients)
with torch.no_grad():
    content_features = model(content_img)
    style_features = model(style_img)

optimizer = optim.Adam([generated_img], lr=0.02)
num_steps = 300

for step in range(num_steps):
    optimizer.zero_grad()

    gen_features = model(generated_img)

    loss = total_loss(
        gen_features,
        content_features,
        style_features,
        generated_img
    )

    loss.backward()
    optimizer.step()

    with torch.no_grad():
        generated_img.clamp_(0, 1)

    if step % 50 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")
        imshow(generated_img, f"Step {step}")
```

```python
# save image

def save_image(tensor, filename):
    image = tensor.clone().detach().cpu().squeeze(0)
    image = image.permute(1, 2, 0)
    plt.imsave(filename, image)

save_image(generated_img, "stylized_output.png")
```
